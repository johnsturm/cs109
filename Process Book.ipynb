## 1.1 Scrape Project Titles

We found this site, Kickspy, from which you can get all the titles of kickstarter projects. In this section, we're scraping from that site and  getting all the titles of projects.

```python
def cat_string(categories):
    s = ""
    for c in categories:
        s += "&Categories=" + str(c)
    return s

def make_url(year, month, day, categories):
    year = str(year)
    month = str(month)
    day = str(day)
    
    return "http://www.kickspy.com/projects/find?Keywords=&Settings.ExpandStatus=true&AnyStatus=true&Settings.ExpandCategories=true" + cat_string(categories) + "&Settings.ExpandDates=true&StartedAfter=" + year + "%2F" + month + "%2F" + day + "&StartedBefore=" + year + "%2F" + month + "%2F" + day + "&SortBy=0"

def get_titles(year, month, day):
    #url = "http://www.kickspy.com/projects/find?Keywords=&Settings.ExpandStatus=true&AnyStatus=true&AllCategories=true&Settings.ExpandDates=true&StartedAfter=" + year + "%2F" + month + "%2F" + day + "&StartedBefore=" + year + "%2F" + month + "%2F" + day + "&SortBy=0"
    
    titles = set()
    
    text = requests.get(make_url(year, month, day, [j for j in range(14)])).text[60928:]
    for line in text.split('\n'):
        if "<h3>" in line:
            l = re.findall(r'">(.*)<\/a><\/h3>', text)
            titles.update([unicodedata.normalize('NFKD', j.strip()).encode('ascii','ignore') for j in l])
        
    text = requests.get(make_url(year, month, day, [j + 14 for j in range(17)])).text[60928:]
    for line in text.split('\n'):
        if "<h3>" in line:
            l = re.findall(r'">(.*)<\/a><\/h3>', text)
            titles.update([unicodedata.normalize('NFKD', j.strip()).encode('ascii','ignore') for j in l])

    text = requests.get(make_url(year, month, day, [j + 31 for j in range(31)])).text[60928:]
    for line in text.split('\n'):
        if "<h3>" in line:
            l = re.findall(r'">(.*)<\/a><\/h3>', text)
            titles.update([unicodedata.normalize('NFKD', j.strip()).encode('ascii','ignore') for j in l])
            
    time.sleep(2)
        
    return titles

def all_titles():
    years = [2009, 2010, 2011, 2012, 2013]
    months = [i + 1 for i in range(12)]
    
    all_titles = []
    
    for y in years:
        for m in months:
            print m
            days = []
            if m in [1, 3, 5, 7, 8, 10, 12]:
                days = [1 + i for i in range(31)]
            elif m in [4,6,9,11]:
                days = [1 + i for i in range(30)]
            else:
                days = [1+i for i in range(28)]
            for d in days:
                all_titles.extend(get_titles(y, m, d))
    
    return all_titles
```

## 1.2 Scrape Project Information

Given a project's title, you can scrape its project information from kickstarter.

This information includes stuff like: project start date, end date, location, goal amount, name of user, title of project, blurb of project, and a whole bunch more. These are most of the fields that we'll want in order to make our predictions.

```python
# here are some special functions (some of them hack-ish) to help us get around a few formatting bugs in the next section of code
# not super interesting

def dict_iterator(my_dict):
    for key, value in my_dict.iteritems():
        if type(value)==list:
            for item in value:
                new_dict = dict_iterator(item)
                my_dict[key] = new_dict
        if type(value)==dict:
            new_dict = dict_iterator(value)
            my_dict[key] = new_dict
        if type(value)==unicode:
            if 'saal_tucker_sturm_zimble' in value:
                value = value.replace('saal_tucker_sturm_zimble','"')
                my_dict[key] = value
    return my_dict

def tm_remover(title):
    words = []
    for word in title.split(' '):
        if 'TM' not in word:
            words.append(word)
    return " ".join(words)
            
def alt_project_fetch(title):
    url_endings = []
    new_dicts = []
    url = 'http://www.kickspy.com/projects/find?Keywords='+title+'&Settings.ExpandStatus=true&AnyStatus=true&AllCategories=true&SortBy=0'
    try:
        text = requests.get(url).text
    except ValueError:
        return None
    for line in text.split('\n'):
        if "0 results found" in line:
            if "TM" in title:
                return alt_project_fetch(tm_remover(title))
            else:
                return None
        elif "<h3>" in line:
            url_ending = line.split('href="',1)[1]
            url_ending = url_ending.split('">',1)[0]
            url_endings.append(url_ending)
    if url_endings==[]:
        return None
    for url_ending in url_endings:
        url_2 = "http://www.kickstarter.com/" + url_ending
        try:
            text_2 = requests.get(url_2).text
        except ValueError:
            return None
        for line in text_2.split('\n'):
            if 'window.current_project' in line:
                data = unicode(html_parser.unescape(line))
                data = data.split('window.current_project = "',1)[1]
                data = data.split('";')[0]
                data = str(data)
                data = data.replace('\\"','\\saal_tucker_sturm_zimble')
                data = data.replace('\\','')
                try:
                    data = json.loads(data)
                except ValueError:
                    continue
                data = dict_iterator(data)
                new_dicts.append(data)
    return new_dicts
```

```python
# function to get info (from kickstarter) about all the projects associated to titles in a list of titles
def get_projects(titles):
    error_titles=[]
    base_url = "http://www.kickstarter.com/projects/search.json"
    fields = ['creator_id', 'creator_name', 'creator_slug',
              'project_id', 'project_name', 'project_slug', 'blurb', 'category',
              'currency', 'goal', 'pledged', 'backers_count',
              'launched_at', 'deadline', 'last_state_change', 'state',
              'location_id', 'location_name', 'country',
              'project_url', 'creator_url']
    projects = dict((field, []) for field in fields)
    
    for i, title in enumerate(titles):
        if i%100 == 0:
            print i
        try:
            data = requests.get(base_url, params = {'search': '', 'term': title}).json()['projects']
        except ValueError:
            data = alt_project_fetch(title)
            if data == None:
                error_titles.append(title)
                continue
            
        if len(data) == 0:
            data = alt_project_fetch(title)
            if data == None:
                error_titles.append(title)
                continue
            data = [data]
            
            
        for project in data:
            if project['id'] not in projects['project_id']:
                
                
                try:
                    projects['creator_id'].append( project['creator']['id'] )
                except KeyError:
                    projects['creator_id'].append( '' )
                    
                try:
                    projects['creator_name'].append( project['creator']['name'].encode('utf-8') )
                except KeyError:
                    projects['creator_name'].append( '' )
                    
                try:
                    projects['creator_slug'].append( project['creator']['slug'].encode('utf-8') )
                except KeyError:
                    projects['creator_slug'].append( '' )
                
                try:
                    projects['project_id'].append( project['id'] )
                except KeyError:
                    projects['project_id'].append( '' )
                    
                try:
                    projects['project_name'].append( project['name'].encode('utf-8') )
                except KeyError:
                    projects['project_name'].append( '' )
                    
                try:
                    projects['project_slug'].append( project['slug'].encode('utf-8') )
                except KeyError:
                    projects['project_slug'].append( '' )
                    
                try:
                    projects['blurb'].append( project['blurb'].encode('utf-8') )
                except KeyError:
                    projects['blurb'].append( '' )
                    
                try:
                    projects['category'].append( project['category']['name'].encode('utf-8') )
                except KeyError:
                    projects['category'].append( '' )
                    
                try:
                    projects['currency'].append( project['currency'].encode('utf-8') )
                except KeyError:
                    projects['currency'].append( '' )
                
                try:
                    projects['goal'].append( project['goal'] )
                except KeyError:
                    projects['goal'].append( '' )
                    
                try:
                    projects['pledged'].append( project['pledged'] )
                except KeyError:
                    projects['pledged'].append( '' )
                
                try:
                    projects['backers_count'].append( project['backers_count'] )
                except KeyError:
                    projects['backers_count'].append( '' )
                    
                try:
                    projects['launched_at'].append( project['launched_at'] )
                except KeyError:
                    projects['launched_at'].append( '' )
                    
                try:
                    projects['deadline'].append( project['deadline'] )
                except KeyError:
                    projects['deadline'].append( '' )
                    
                try:
                    projects['last_state_change'].append( project['state_changed_at'] )
                except KeyError:
                    projects['last_state_change'].append( '' )
                    
                try:
                    projects['state'].append( project['state'].encode('utf-8') )
                except KeyError:
                    projects['state'].append( '' )
                    
                try:
                    projects['location_id'].append( project['location']['id'] )
                except KeyError:
                    projects['location_id'].append( '' )
                    
                try:
                    projects['location_name'].append( project['location']['name'].encode('utf-8') )
                except KeyError:
                    projects['location_name'].append( '' )
                    
                try:
                    projects['country'].append( project['country'].encode('utf-8') )
                except KeyError:
                    projects['country'].append( '' )
                
                try:
                    projects['project_url'].append( project['urls']['web']['project'].encode('utf-8') )
                except KeyError:
                    projects['project_url'].append( '' )
                
                try:
                    projects['creator_url'].append( project['creator']['urls']['web']['user'].encode('utf-8') )
                except KeyError:
                    projects['creator_url'].append( '' )
            
    return error_titles, pd.DataFrame(projects)
```

```python
#this function outputs a modified copy of an inputted data frame in which the 
#values in the selected fields are converted from timestamps to the string-version of datetimes

def time_to_datetimes(input_df, fields):
    df = input_df.copy()
    for field in fields:
        df[field] = df[field].astype(str)
        for i, time in enumerate(df[field]):
            df[field][i] = str(datetime.datetime.fromtimestamp(float(time)))
    return df
```

## 1.3 Scrape User Data

Given a user's url, you can scrape his/her information from kickstarter.

This information includes stuff like: join date, how many projects he/she created, how many projects he/she backed, location, facebook link, twitter link, and a whole bunch more. These contain a few of the fields for the predictions. It had to do two get request per user (about 200k overall) and thus ran pretty slowly, so we split the process amongst ourselves and then rejoined the csvs in excel.

```python
def get_user_data(titles_df):    
    fields = ['creator_name', 'creator_slug',
              'projects_created', 'projects_backed','creator_id', 'location', 'date_joined',
              'facebook_link','twitter_link', 'creator_url']
    individual_users = dict((field, []) for field in fields)
    missing_created_project_data = []
    missing_backed_project_data = []
    url_list = all_projects['creator_url'].unique()
    
    #For John
    for i,url in enumerate(url_list[25986: 51972]):
    
    #For Eli
    #for i,url in enumerate(url_list[51972: 77958]):
    
    #For Aaron
    #for i,url in enumerate(url_list[77958: ]):
    
    #For Jake
    #for i,url in enumerate(url_list[: 25986]):
        if i%1000 == 0:
            print i
        # Define Variables
        projects_created_name = []
        projects_created_id = []
        projects_created_missing = []
        projects_backed_url = []
        projects_backed_id = []
        projects_backed_missing = []
        projects_created = ''
        projects_backed = ''
        location = ''
        date_joined = ''
        facebook_link = ''
        twitter_link = ''

        # Get userpage data
        usr_df = titles_df[titles_df.creator_url == url]
        xml = requests.get(url).text
        dom = web.Element(xml)
        
        # Get data from project dataframe
        creator_slug = usr_df.creator_slug.unique()
        creator_slug = creator_slug[0]
        creator_name = usr_df.creator_name.unique()
        creator_name = creator_name[0]
        creator_id = usr_df.creator_id.unique()
        creator_id = creator_id[0]

        # Get the projects the user created
        for project in usr_df.project_name:
            project = project.decode('utf-8')
            projects_created_name.append(project)
        for subset in dom.by_class("profile_created_projects"):
            for item in subset.by_class("project_name"):
                project = item.content.replace('\n','')
                project = unicode(html_parser.unescape(project))
                if project not in projects_created_name:
                    projects_created_missing.append( project )
                    missing_created_project_data.append( project )
        for project_id in usr_df.project_id:
            project_id = int( project_id )
            projects_created_id.append( str(project_id) )
        projects_created = ','.join(projects_created_id)
        
        
        # Get the projects the user backed
        for subset in dom.by_class("mobius"):
            for subdict in subset.by_tag('a'):
                base_url = 'http://www.kickstarter.com'
                projects_backed_url.append( base_url + subdict.attrs['href'] )
        for project_backed_url in projects_backed_url:
            backed_id = titles_df.project_id[titles_df.project_url == project_backed_url]
            if len(backed_id) == 0:
                backed_id = titles_df.project_id[titles_df.project_url == project_backed_url+'?ref=live']
            if len(backed_id) == 0:
                backed_id = 0
                missing_backed_project_data.append( project_backed_url )
            backed_id = backed_id.iloc[0,:]
            projects_backed_id.append( str(backed_id) )
        projects_backed = ','.join(projects_backed_id)
    
        #Get Location
        for city in dom.by_class("location"):
            city = city.content.replace('\n','')
            city = unicode(html_parser.unescape(city))
            location = city.encode('utf-8')
    
        # Get User Join Date
        for join_date in dom.by_class("joined"):
            join_date = join_date.content.replace('\n','')
            join_date = join_date.replace('Joined','')
            join_date = unicode(html_parser.unescape(join_date))
            date_joined = join_date.encode('utf-8')
            
        #Access A Seperate Dom for the Bio Data    
        for line in dom.by_class('remote_modal_dialog'):
            bio_url = 'http://www.kickstarter.com'+ line.attrs['href']
            bio_xml = requests.get(bio_url).text
            bio_dom = web.Element(bio_xml)
            for subset in bio_dom.by_class('menu-submenu'):
                for link in subset('a'):
                    my_link = link.attrs['href']
                    if 'facebook' in my_link:
                        facebook_link = my_link
                    if 'twitter' in my_link:
                        twitter_link = my_link
        
        #Update the Dictionary
        individual_users['creator_url'].append( url )
        individual_users['creator_slug'].append( creator_slug )
        individual_users['creator_name'].append( creator_name )
        individual_users['creator_id'].append( creator_id )
        individual_users['projects_created'].append( projects_created )                
        individual_users['projects_backed'].append( projects_backed )
        individual_users['location'].append( location )
        individual_users['date_joined'].append( date_joined )
        individual_users['facebook_link'].append( facebook_link )
        individual_users['twitter_link'].append( twitter_link )
        
    return missing_created_project_data, missing_backed_project_data, pd.DataFrame(individual_users)
```

## 1.5 Combine User and Project Data

In order to use the user data, we had to add the information to the projects dataframe. This code does that and outputs a large dataframe with all the scraped data.

```python
def update_projects_with_users(project_df, user_df):
    project_df = project_df.copy()
    user_df = user_df.copy()
    project_df['facebook'] = project_df['creator_id'].map(lambda x: 0)
    project_df['twitter'] = project_df['creator_id'].map(lambda x: 0)
    project_df['creator_location'] = project_df['creator_id'].map(lambda x: 'string')
    project_df['creator_joined'] = project_df['creator_id'].map(lambda x: 'string')
    for i in range(len(project_df)):
        creator_join_date = user_df.date_joined[user_df.creator_id == project_df.creator_id[i]]
        creator_join_date = creator_join_date.iloc[0]
        if type(creator_join_date) != type(0.):
            creator_join_date = creator_join_date.split('"',2)[1]
            if 'T' in creator_join_date:
                date, time = creator_join_date.split('T',)
                time = time.split('-',)[0]
                creator_join_date = date+' '+time
                creator_join_date = pd.to_datetime(creator_join_date)
                project_df.creator_joined[i] = creator_join_date
            else:
                project_df.creator_joined[i] = creator_join_date
        location = user_df.location[user_df.creator_id == project_df.creator_id[i]]
        location = location.iloc[0]
        project_df.creator_location[i] = location
        twitter = user_df.twitter_link[user_df.creator_id == project_df.creator_id[i]]
        twitter = twitter.iloc[0]
        facebook = user_df.facebook_link[user_df.creator_id == project_df.creator_id[i]]
        facebook = facebook.iloc[0]
        if type(facebook) == type(0.):
            project_df.facebook[i] = 0
        else:
            project_df.facebook[i] = 1
        if type(twitter) == type(0.):
            project_df.twitter[i] = 0
        else:
            project_df.twitter[i] = 1
        if i%1000 == 0:
            print i
    return project_df
```

```python
def convert_to_dollars(project_df):
    converted_value = []
    for i in range(len(project_df)):
        if project_df.currency[i] == 'USD':
            converted_value.append(project_df.goal[i])
        elif project_df.currency[i] == 'CAD':
            convert_goal = float(project_df.goal[i]*0.94)
            converted_value.append(convert_goal)
        elif project_df.currency[i] == 'GBP':
            convert_goal = float(project_df.goal[i]*1.64)
            converted_value.append(convert_goal)
        elif project_df.currency[i] == 'NZD':
            convert_goal = float(project_df.goal[i]*0.83)
            converted_value.append(convert_goal)
        elif project_df.currency[i] == 'AUD':
            convert_goal = float(project_df.goal[i]*0.92)
            converted_value.append(convert_goal)
    project_df['goal_dollars'] = converted_value
    return project_df
```

```python
# Defining Cost Functions
def log_likelihood(clf, x, y):
    """ The log likelihood cost function for each prediction is:
            the log of the probability of success if the project succeeded
            or the log of the probability of failure if the project failed
        And for the whole classifier, is the sum of the cost of each prediction
    """
    logPs = clf.predict_log_proba(x)
    return sum([logPs[i][1] for i in range(len(y)) if y[i] == 1]) + sum([logPs[i][0] for i in range(len(y)) if y[i] == 0])

def fraction_correct(clf, x, y):
    logPs = clf.predict_log_proba(x)
    guess = np.around(logPs)
    return np.mean(guess == y)

# Utilities for splitting up the dataframes into test and train sets
def gen_random(val):
    # This is a helper function for disjoint_sets
    if random.random() < val:
        return False
    else:
        return True
    
def disjoint_sets(length, proportion=.25):
    """This function selects the indices for a test set and a training set that do not overlap, and which together
        take up the entire set of some given dataframe
        
        It returns sets of 0s and 1s describing wither or not the ith index should be included
    """
    include = [gen_random(proportion) for i in range(length)]
    exclude = [not(i) for i in include]
    return include, exclude

def prelim_split(df, field, proportion=.25):
    """
    This function uses disjoint_sets to get a training set and a test set, then returns panda Series for the
    training data's x and y fields and the test data's x and y fields. It returns the value of the column of the
    dataframe, and vectorizing the x happens elsewhere -- we want each model to remain free to choose its own
    min_df.
    
    It also returns x and y, so that x can be used to fit the vectorizer, so that the different models can still
    predict based on words that aren't necessarily in its set of the data.
    
    X SHOULD NEVER TOUCH THE CLASSIFIER, ONLY THE VECTORIZER
    """
    # This section splits between train and test sets
    y='success'
    i, e = disjoint_sets(len(df), proportion)
    train = df[i]
    test = df[e]
    
    # Get X and Y
    x = df[field]
    y = df['success']
    
    # Get the training and test data for the x and y
    xtrain = train[field]
    xtest_final = test[field]
    ytrain = train['success']
    ytest_final = test['success']
    
    return x, y, xtrain, ytrain, xtest_final, ytest_final

def make_xy(x, y, vectorizer=None):
    """
    This function vecotrizes some x, and returns x and y
    """
    text = x
    
    if vectorizer == None:
        vectorizer = CountVectorizer(min_df=0)

    vectorizer.fit(text)
    x = vectorizer.transform(text)
    #X = X.toarray()
    
    return x, y

# Utilities for plotting the predictions/calibration of a model
def ysigma(frac, n):
    # This just returns the sigma for y for plotting the calibration plpot
    return math.sqrt(frac*(1.0 - frac)/n)

def calibration_plot(clf, X, Y):
    """
    This shows the calibration plot for a classifier, taking the predictions, sorting them into groups based on
    predicted success rates, and then aggregating the actual success rate of that group
    """
    # Get the probabilities, set the number of bins
    nbins = 30
    proba = [row[1] for row in clf.predict_proba(X)]
    
    # make a dataframe for predicted probabilities and observed success, and sort them into bins
    df        = pd.DataFrame({"proba": proba, "label": Y})
    df["bin"] = df.apply(lambda x: math.floor(x['proba'] * nbins), axis=1)
    
    # split the dataframe into groups based on what section of the probability distribution it falls in
    # then, aggregate it to get the observed success rates
    bins      = df.groupby("bin")
    observed  = bins['label'].aggregate(np.mean).values
    predicted = bins['proba'].aggregate(np.mean).values
    bin_size  = bins['label'].aggregate(len).values
    
    # calculate sigmas, spread it out within two standard deviations
    sigmas    = [2*ysigma(observed[i], bin_size[i]) for i in range(len(bin_size))] # 95% confidence interval
    
    # plot the observed vs the predicted with error bars, and a line for perfect calibration
    plt.errorbar(predicted, observed, yerr=np.vstack([sigmas, sigmas]))
    plt.plot([0,1], [0,1], color="r")
    plt.show()
    
    # plot the distribution of probability assignments
    plt.hist(proba, color='orange')
    plt.show()

def train_score_display(vectorizer, clf, x, xtrain, ytrain, xtest, ytest, fit_params={}):
    """
    This function fits a model's vectorizer to all the data (so that it can handle arbitrary data input),
    fits the classifier to x values vectorized using that vectorizer, then does a calibration plot,
    returning the resulting vecotrizer and classifier
    
    Also, it never gets passed y so that no modification from within it lets it train on the whole set
    """
    # Fit and transform according to the vectorizer
    vectorizer.fit(x)
    xtrain_v = vectorizer.transform(xtrain)
    xtest_v = vectorizer.transform(xtest)
    
    # train it
    if 'weight' in fit_params:
        clf.fit(xtrain_v, ytrain, weight=fit_params['weight'])
    else:
        clf.fit(xtrain_v, ytrain)
    
    # score and plot it
    score = clf.score(xtest_v, ytest)
    print "Score: ", score
    calibration_plot(clf, xtest_v, ytest)
    
    return vectorizer, clf

# This is the code for the random exploration of the parameter space
def mult_dict(d, s):
    """This function multiplies every value in a dictionary by a scalar"""
    ans = {}
    for key in d.keys():
        ans[key] = d[key] * s
    return ans

def test_precision(params, sigmas, precision):
    """
    This function checks whether the 95% confidence interval around a set of parameters is within precision proportion
    of the current parameters.
    
    I.e. if precision = .1, it checks the 95% confidence interval falls within +- 5% of each parameter
    """
    for key in params.keys():
        target = params[key] * precision / 2.0
        if sigmas[key] * 2 > target:
            return False
    return True
        
def random_move(params, sigmas):
    """
    This function has a 50% chance of taking a normally distributed step from the current position along each parameter
    """
    ans = {}
    for key in params.keys():
        if random.random() < .5: # do we change this parameter?
            ans[key] = -1
            while ans[key] <= 0: # keep the parameter above zero
                ans[key] = params[key] + random.gauss(0, sigmas[key])
        else:
            ans[key] = params[key]
            
    if params == ans: # if no changes, try again
        return random_move(params, sigmas)
    return ans

def random_optimization(model, params, sigmas, xtrain, ytrain, precision = .1):
    """
    This function implements random_optimization, taking normally distributed steps along each parameter, and keeping
    the change if it's an improvement. When the current best position remains the best, decrease sigma to take
    smaller steps.
    """
    best_params = params
    i = 0
    best_score  = model(best_params, xtrain, ytrain) # look at the best score
    
    while not test_precision(best_params, sigmas, precision): # repeat until we achive the desired precision
        new_params = random_move(best_params, sigmas)    # take a step
        new_score  = model(new_params, xtrain, ytrain)   # score the new model
        
        if new_score > best_score:   # if it's better, trade for it
            best_params = new_params
            best_score  =  new_score
        else:                        # otherwise, decrease sigma
            sigmas = mult_dict(sigmas, .9)
        
        i += 1
        if i % 10 == 1:
            print i, best_params, best_score, sigmas
            
    return best_params, best_score

# These functions all make a vectorizer and classifier with the given parameters, then do a cv score on the training set
def nb_classifier(params, xtrain, ytrain):
    vectorizer = CountVectorizer(min_df = params['min_df'])
    vectorizer.fit(xtrain)
    X, Y = make_xy(xtrain, ytrain, vectorizer)
    clf = MultinomialNB(alpha = params['alpha'])
    return np.mean(cv.cross_val_score(clf, X, Y, scoring=log_likelihood))

def log_classifier(params, xtrain, ytrain):
    vectorizer = CountVectorizer(min_df = params['min_df'])
    vectorizer.fit(xtrain)
    X, Y = make_xy(xtrain, ytrain, vectorizer)
    clf = sklm.LogisticRegression(C=params['C'])
    return np.mean(cv.cross_val_score(clf, X, Y, scoring=log_likelihood))

def dt_classifier(params, xtrain, ytrain):
    vectorizer = CountVectorizer(min_df = params['min_df'])
    vectorizer.fit(xtrain)
    X, Y = make_xy(xtrain, ytrain, vectorizer)
    clf =  sktr.DecisionTreeClassifier()
    return np.mean(cv.cross_val_score(clf, X.toarray(), Y, scoring=tree_cost))
```

```python
# function to plot distribution of goals:
def distr_of_lengths(df, field):
    data = [len(str(text)) for text in df[field]]
    
    plt.hist(data, bins=20)
    plt.title('Distribution of ' + field + ' length')
    plt.xlabel('character count')
    remove_border()
    plt.show()
    
# function to plot distribution of goals:
def distr_of_caps(df, field):
    data = [np.mean([x.isupper() for x in str(text)]) for text in df[field]]
    
    plt.hist(data, bins=20)
    plt.title('Distribution of percent capital letters in' + field )
    plt.xlabel('percent caps')
    plt.semilogy()
    remove_border()
    plt.show()
    
# function to plot distribution of goals:
def distr_of_word_lengths(df, field):
    data = [np.mean([len(word) for word in str(text).split(' ')]) for text in df[field]]
    
    plt.hist(data, bins=40)
    plt.title('Distribution of average word lengths in' + field )
    plt.xlabel('average word length')
    plt.semilogy()
    remove_border()
    plt.show()

# function to plot success vs percent caps of a text field
def success_by_caps(input_df, field):
    
    df = input_df[input_df['state'] != 'live'].reset_index()
    caps_dict = {}
    
    for i, text in enumerate(df[field]):
        caps = round(np.mean([x.isupper() for x in str(text)]),2)
        
        if caps not in caps_dict.keys():
            caps_dict[caps] = {'successful': 0., 'total': 0.}
                                 
        if df['state'][i] == 'successful':
            caps_dict[caps]['successful'] += 1.
        caps_dict[caps]['total'] += 1.
        
    options = caps_dict.keys()
    data = [caps_dict[caps]['successful'] / caps_dict[caps]['total'] for caps in options]
                
    plt.plot(options, data, 'ro')
    plt.xlabel('fraction capitalized in' + field)
    plt.ylabel('fraction successful')
    plt.title('Success by fraction capitalized in ' + field)
    remove_border()
    plt.show()
    
# function to plot success vs avg word length of a text field
def success_by_avglength(input_df, field):
    
    df = input_df[input_df['state'] != 'live'].reset_index()
    caps_dict = {}
    
    for i, text in enumerate(df[field]):
        caps = np.mean([len(word) for word in str(text).split(' ')])
        caps = round(caps,0)
        
        if caps not in caps_dict.keys():
            caps_dict[caps] = {'successful': 0., 'total': 0.}
                                 
        if df['state'][i] == 'successful':
            caps_dict[caps]['successful'] += 1.
        caps_dict[caps]['total'] += 1.
        
    options = caps_dict.keys()
    data = [caps_dict[caps]['successful'] / caps_dict[caps]['total'] for caps in options]
                
    plt.plot(options, data, 'ro')
    plt.xlabel('avg word length' + field)
    plt.ylabel('fraction successful')
    plt.title('Success by fraction avg word length in ' + field)
    remove_border()
    plt.show()
    
    
# function to plot success vs length of a field
def success_by_length(input_df, field):
    
    df = input_df[input_df['state'] != 'live'].reset_index()
    length_dict = {}
    
    for i, text in enumerate(df[field]):
        length = len(str(text))
        length = length - ( length % 5 )
        
        if length not in length_dict.keys():
            length_dict[length] = {'successful': 0., 'total': 0.}
                                 
        if df['state'][i] == 'successful':
            length_dict[length]['successful'] += 1.
        length_dict[length]['total'] += 1.
        
    lengths = length_dict.keys()
    data = [length_dict[length]['successful'] / length_dict[length]['total'] for length in lengths]
                
    plt.plot(lengths, data, 'ro')
    plt.xlabel('length of ' + field)
    plt.ylabel('fraction successful')
    plt.title('Sucess by length of ' + field)
    remove_border()
    plt.show()
```

## 3.1 Models for numerical data

### 3.1.1 - Framework

The goal here is to write code that will take in the projects and optimally fit a curve to describe the effect a numerical field - like blurb length, project goal, etc. - has on chance of success.

The input to this code will be a dataframe of projects and a function obtaining some field for each such project.

The output from this code will be a function that takes a project and predicts its success based on the field of interest. The function outputted will minimize something like log-likelihood, or number of correct guesses, etc.

There are many different classifiers (thank you, sklearn) that we could use to make these models. How to choose:

(1) We know from our exploration that some of the curves we're trying to model have strange-looking or jagged-y. This suggests that doing polynomial models might be messy.

(2) None of our numerical fields are significantly correlated (section 2.5, 2.6) so we can model them independently. This means that our data is going to be quite dense - we have many thousand points, and only one curve to infer from it.

Conclusion - kNN approaches should handle the jagged-ness well, since there is no underlying model. And we don't have to worry about them having sparcity problems since we're only working in 1D. So let's try kNN over a bunch of k's. We'll also through in a logistic regression for the heck of it and see it it ever outperforms kNN.

Finally, just an overview of what we'll have our function do here (for each numerical field): 

(1) do a recap in graphs of the explanatory section. Note that OUR HISTOGRAMS PLOT WITH POINTS, NOT BARS (since plt.hist has trouble with log scales) 

(2) On some random subset of the training data, go through several different classifiers and obtain a score via kfold cross-validation. 

(3) Pick the model with the best log-likelihood score, plot what its predictions look like.

(4) Print out how well the model performs on the test data (which we're NOT using to optimize the model), according to log-likelihood and fraction-guessed-correctly (by rounding estimate).

A Final Note:

We're not using true log-likelihood. Rather we've using a version that penalizes impossible guesses as less than infinitely bad. It makes sense to do this - that is, to raise 0 probability guesses to 1/1000 and lower 1 probability guesses to 999/1000 - because we don't intend for our model to be accurate to 0.1% likelihood, so it wouldn't make sense to penalize it for being off this much. Another way of looking at this is that we're changing our classifier to avoid making 0 or 1 guesses.

```python
#first some functions to give an explanatory recap

#doing this one the hard way since plt histograms are junk with log scales (similar to success_by_field)
def numerical_distr(field_name, values, log_x, log_y):
    num_buckets = 30
    min_value, max_value = min(values), max(values)
    buckets = [0 for i in xrange(num_buckets)]
    
    if log_x:
        bucket_space = np.logspace(np.log(min_value) / np.log(10.), np.log(max_value) / np.log(10.), num_buckets)
    else:
        bucket_space = np.linspace(min_value, max_value, num_buckets)
        
    for i, value in enumerate(values):
        num = 0
        for j, term in enumerate(bucket_space):
            if value < term:
                num = j - 1
                break
        buckets[num] += 1
        
    plt.plot(bucket_space, buckets, 'ro')
    plt.xlabel(field_name)
    if log_x:
        plt.semilogx()
    if log_y:
        plt.semilogy()
    plt.title('Number of projects by ' + field_name)
    remove_border()
    plt.show() 

    
def success_by_field(field_name, values, successes, log_x):
    
    num_buckets = 50
    min_value, max_value = min(values), max(values)
    buckets = [{'successful': 0., 'total': 0.} for i in xrange(num_buckets)]
    
    if log_x:
        bucket_space = np.logspace(np.log(min_value) / np.log(10.), np.log(max_value) / np.log(10.), num_buckets)
    else:
        bucket_space = np.linspace(min_value, max_value, num_buckets)
        
    for i, value in enumerate(values):
        num = 0
        for j, term in enumerate(bucket_space):
            if value < term:
                num = j - 1
                break         
        buckets[num]['successful'] += 1 if successes[i] else 0
        buckets[num]['total'] += 1 
    
    data = [bucket['successful'] / (bucket['total'] + 0.01) for bucket in buckets]
    
    plt.plot(bucket_space, data, 'ro')
    plt.xlabel(field_name)
    plt.ylabel('fraction successful')
    if log_x:
        plt.semilogx()
    plt.title('Success by ' + field_name)
    remove_border()
    plt.show() 

# now we want to fit some curves to our model
# this is our starting cost functions
def log_likelihood(clf, x, y):
    L = 0
    proba = clf.predict(x)
    
    #fudged slightly so that things don't get wrecked by zero-probability events (see explanation in markdown above)
    for i, term in enumerate(proba):
        proba[i] += 1./1000 if term == 0 else 0
        proba[i] -= 1./1000 if term == 1 else 0
        
    true_logs = np.log(proba)
    false_logs = np.log(1 - proba)
    
    for index, value in enumerate(y):
        L += true_logs[index] if value == 1 else false_logs[index]
            
    return L

def fraction_correct(clf, x, y):
    proba = clf.predict(x) 
    guess = np.around(proba)
    return np.mean(guess == y)

def both_metrics(clf, x, y):
    L = 0
    proba = clf.predict(x)
    
    #fudged slightly so that things don't get wrecked by zero-probability events
    for i, term in enumerate(proba):
        proba[i] += 1./1000 if term == 0 else 0
        proba[i] -= 1./1000 if term == 1 else 0
        
    true_logs = np.log(proba)
    false_logs = np.log(1 - proba)
    
    for index, value in enumerate(y):
        L += true_logs[index] if value == 1 else false_logs[index]
    
    guess = np.around(proba)
    fraction_correct = np.mean(guess == y)
    
    return [L, fraction_correct]
        

#this cross validates with diff random subsets of the data
def cv_score(clf, x, y, score_func):
    result = 0
    nfold = 5
    for train, test in KFold(y.size, nfold): # split data into train/test groups, 5 times
        clf.fit(x[train], y[train]) # fit
        result += score_func(clf, x[test], y[test]) # evaluate score function on held-out data
    return result / nfold # average

#for making my own classifiers that fit into this framework:
#Smoothing Classifier works like kNN but if there are points with same x-value, includes all
class SmoothingClassifier():
    def __init__(self, min_group):
        self.x = np.array([])
        self.y = np.array([])
        self.clf = []
        self.min_group = min_group
        
    def fit(self, X, Y):
        clf = KNeighborsRegressor(n_neighbors=self.min_group, weights='uniform')
        clf.warn_on_equidistant = False
        clf.fit(X, Y)
        
        self.clf = clf
        self.x = X
        self.y = Y
        
    def predict(self, X):
        
        count = 0
        
        result = self.clf.predict(X)
        for i, x in enumerate(X):
            matches = np.array(np.where(self.x == x))
            if matches.size > self.min_group:
                count += 1
                result[i] = np.mean(self.y[matches])
        
        print count
        
        return result

#this generates a bunch of different classifiers we'll try:
def generate_classifiers():
    result = []
    
    for k in [3**(n+1) for n in xrange(6)]:      #xrange(8)]: sometimes these are the best models, but they take too long to run
        #Skip now
        #clf = SmoothingClassifier(k)
        #result.append(clf)
        
        clf = KNeighborsRegressor(n_neighbors=k, weights='uniform', algorithm='auto') #uniform since distance -> division by zero
        clf.warn_on_equidistant = False
        result.append(clf)
        
    clf = LogisticRegression(C=1.0, penalty='l1')
    result.append(clf)
    
    return result

# this tests out a bunch of possible classifiers
def optimize_classifier(x, y, to_max):
    best_score, best_clf = - np.inf, []
    
    for clf in generate_classifiers():
        score = cv_score(clf, x, y, to_max)
        
        #print clf
        #print score
        #print
                 
        if score > best_score:
            best_score = score
            clf.fit(x, y)
            best_clf = clf
    
    return best_score, best_clf

# plot a rough outline of the predictions our model makes
def plot_classifier(clf, x, score, field_name, log_x):
    open_x = [term[0] for term in x]
    
    if log_x:
        pre_x_plot = np.logspace(np.log(min(open_x)) / np.log(10.), 
                                 np.log(max(open_x)) / np.log(10.), 
                                 1000)
    else:
        pre_x_plot = np.linspace( min(open_x), max(open_x), 1000)
    x_plot = np.array([[term] for term in pre_x_plot])
    
    predictions = clf.predict(x_plot)
    
    plt.scatter(pre_x_plot, predictions)
    plt.title('prediced project success vs ' + field_name + ' (score: ' + str(score) + ')')
    plt.xlabel(field_name)
    plt.xlim([min(pre_x_plot), max(pre_x_plot)])
    plt.ylim([min(0, min(predictions)), max(1, max(predictions))])

    if log_x:
        plt.semilogx()

    plt.show()

# this is the total thing that we'll run - starting with the training and test values in a field and whether or not
# the corresponding projects succeeded, we print out explanatory graphs and we optimize a classifier 
def numerical_model(train_values, train_successes, test_values, test_successes, name, to_max, training_size, log_x = False, distr_log_y = False):
    
    #format
    train_values, train_successes = np.array(train_values), np.array(train_successes)
    
    #get an appropriately size random subset of the data to train with so code doesn't take forever
    random_indices = (np.random.random(train_successes.size) < (float(training_size) / train_successes.size))
    
    values, successes = np.array([]), np.array([])
    for i, value in enumerate(random_indices):
        if value:
            values = np.append(values, np.array([train_values[i]]))
            successes = np.append(successes, np.array([train_successes[i]]))
    
    #plot the distribution of this field and its successes
    numerical_distr(name, values, log_x, distr_log_y)
    success_by_field(name, values, successes, log_x)
    
    #find the best model and plot the results
    values = np.array([[term] for term in values])
    score, clf = optimize_classifier(values, successes, to_max)
    plot_classifier(clf, values, score, name, log_x)
    print 'best' + str(clf)
    
    metrics = both_metrics(clf, np.array([[term] for term in test_values]), test_successes)
    print 'log_likelihood: ' + str(metrics[0])
    print 'fraction guessed correctly: ' + str(metrics[1])
    
    return clf.predict(np.array([[term] for term in train_values])), clf.predict(np.array([[term] for term in test_values]))
```

## 3.2 - Models for Categorical Data

### 3.2.1 - Framework

Here we're trying to do exactly what we did in the numerical case, just for categorical rather than numerical data - that is, given a numerical field, we want to print out some recap about its distribution and relationship with success, find the best (according to log-likelihood) model we can for it, and then save the predictions that the model makes on our training set (for the model combination later on).

A very convenient fact: for categorical data, log-likelihood is easy to minimize. Assuming that we have enough data in all categories to outweigh any priors we might have about how they affect success (and we almost always do have enough), then the best prediction is to say that a project with value A in category B will succeed with probability equal to the fraction of projects in the training set that had value A that succeeed. This also maximizes the fraction-guessed metric. So the model-making is very easy!

Functions to remind ourselves of the bargraphs from above

```python
# plot a pretty histograph of the distribution of projects according to some categories
def hist_categorical_dist(full_array,title):
    my_dict = defaultdict(int)
    for word in full_array:
        my_dict[word] += 1
    values = []
    labels = []
    sorted_dict = sorted(my_dict.iteritems(), key=operator.itemgetter(1))
    for label,value in sorted_dict:
        labels.append(label)
        values.append(value)
    lefts = []
    locs = []
    colors = []
    for i in xrange (len(labels)):
        lefts.append(i*2+.1)
        locs.append(i*2+1)
        if i%2 == 0:
            colors.append('c')
        else:
            colors.append('b')
    fig = plt.figure()
    fig.set_size_inches(20.,5.)
    y_max = np.max(values)+1000
    plt.bar(left=lefts,height=values,width=1.8,color= colors)  
    plt.xlim(0,2*len(labels))
    plt.ylim(0,y_max)
    plt.xticks(locs,labels,rotation=90,fontsize=8)
    remove_border()
    plt.title(title)
    plt.show()
"""
Function
--------
success_by_category:
    Plots a bar graph of success percent by category

Assumptions
-----------
One or more successes per category

Parameters
-----------
full_array : Numpy ndarray
    ndarray of all data for categories
success_array : Numpy ndarray
    subset of the full array for project categories if the project was successful
title: string
    title of the bargraph to be printed
    
Returns
-------
None

Print Outs:
-----------
Bar graph of percent success by category
"""    

def success_by_category(full_array,success_array,title):
    my_dict_1 = defaultdict(int)
    for word in full_array:
        my_dict_1[word] = 0
    for word in success_array:
        my_dict_1[word] += 1
    my_dict_2 = defaultdict(int)
    for word in full_array:
        my_dict_2[word] += 1    
    values_1 = []
    labels_1 = []
    values_2 = []
    labels_2 = []
    values = []
    labels = []
    tuples = []
    for label,value in my_dict_1.items():
        labels_1.append(label)
        values_1.append(value)
    sorted_dict = sorted(my_dict_2.iteritems(), key=operator.itemgetter(1))
    for label,value in sorted_dict:
        labels_2.append(label)
        values_2.append(value)
    for i in xrange(len(labels_2)):
        for j in xrange(len(labels_1)):
            if labels_2[i] == labels_1[j]:
                my_tup  = (float(values_1[j])/values_2[i],labels_1[j])
                tuples.append(my_tup)
    for i in xrange(len(tuples)):
        values.append(tuples[i][0])
        labels.append(tuples[i][1])
    lefts = []
    locs = []
    colors = []
    for i in xrange (len(labels)):
        lefts.append(i*2+.1)
        locs.append(i*2+1)
        if i%2 == 0:
            colors.append('c')
        else:
            colors.append('b')
    fig = plt.figure()
    fig.set_size_inches(20.,5.)
    plt.bar(left=lefts,height=values,width=1.8,color= colors)  
    plt.xlim(0,2*len(labels))
    plt.ylim(0,1)
    plt.xticks(locs,labels,rotation=90,fontsize=8)
    remove_border()
    plt.title(title)
    plt.show()

# Here we're just writing our own classifier that models categorical data as we discussed earlier
# namely, the probability of success it gives is just the rate of success of things in some category in the fitting set
class Classifier():
    def __init__(self):
        self.proba_dict = {}
        self.average = 0
        
    def fit(self, X, Y):
        grouper = {}
        for i, term in enumerate(X):
            if term not in grouper.keys():
                grouper[term] = {'successful': 0., 'total': 0.}
            if Y[i]:
                grouper[term]['successful'] += 1.
            grouper[term]['total'] += 1.
        
        self.proba_dict = dict((key, float(grouper[key]['successful']) / grouper[key]['total']) for key in grouper.iterkeys())
        self.average = np.mean(Y)
    def predict(self, X):
        
        result = []
        for term in X:
            try:
                result.append(self.proba_dict[term])
            except KeyError:
                result.append(self.average)
        
        return np.array(result)
    
# here are our two metrics for evaluating how good predictions are
def log_likelihood(clf, x, y):
    L = 0

    proba = clf.predict(x)
    #fudged slightly so that things don't get wrecked by zero-probability events
    for i, term in enumerate(proba):
        proba[i] += 1./1000 if term == 0 else 0
        proba[i] -= 1./1000 if term == 1 else 0
    
    for index, value in enumerate(y):
        L += np.log(proba[index]) if value == 1 else np.log(1 - proba[index])
   
    return L 

def fraction_correct(clf, x, y):
    proba = clf.predict(x)
    guess = np.array([round(term, 0) for term in proba])
    return np.mean(guess == y)

# putting it all together
# this function that takes in training and testing data frames and te column we're interested in
# then it outputs some explanatory graphs and the model's predictions
def categorical_model(train_df,test_df,grouping,title):
    if len(set(list(train_df[grouping])))<=100:
        hist_categorical_dist(train_df[grouping],title)
        success_by_category(train_df[grouping],train_df[grouping][train_df.state == 'successful'],title)
    clf = Classifier()
    x_set_train = np.array(train_df[grouping])
    x_set_test = np.array(test_df[grouping])
    y_set_train = np.array([int(term == 'successful') for term in train_df.state])
    y_set_test = np.array([int(term == 'successful') for term in test_df.state])
    clf.fit(x_set_train, y_set_train)
    l = log_likelihood(clf, x_set_test, y_set_test)
    frac_correct = fraction_correct(clf, x_set_test, y_set_test)
    print grouping+' Log Likelihood Score: ' + str(l)
    print grouping+' Fraction Guess Correct: ' + str(frac_correct)
    return clf
    
```

```python
# Defining Cost Functions
def log_likelihood(clf, x, y):
    """ The log likelihood cost function for each prediction is:
            the log of the probability of success if the project succeeded
            or the log of the probability of failure if the project failed
        And for the whole classifier, is the sum of the cost of each prediction
    """
    logPs = clf.predict_log_proba(x)
    return sum([logPs[i][1] for i in range(len(y)) if y[i] == 1]) + sum([logPs[i][0] for i in range(len(y)) if y[i] == 0])

def fraction_correct(clf, x, y):
    logPs = clf.predict_log_proba(x)
    guess = np.around(logPs)
    return np.mean(guess == y)

# Utilities for splitting up the dataframes into test and train sets
def gen_random(val):
    # This is a helper function for disjoint_sets
    if random.random() < val:
        return False
    else:
        return True
    
def disjoint_sets(length, proportion=.25):
    """This function selects the indices for a test set and a training set that do not overlap, and which together
        take up the entire set of some given dataframe
        
        It returns sets of 0s and 1s describing wither or not the ith index should be included
    """
    include = [gen_random(proportion) for i in range(length)]
    exclude = [not(i) for i in include]
    return include, exclude

def prelim_split(df, field, proportion=.25):
    """
    This function uses disjoint_sets to get a training set and a test set, then returns panda Series for the
    training data's x and y fields and the test data's x and y fields. It returns the value of the column of the
    dataframe, and vectorizing the x happens elsewhere -- we want each model to remain free to choose its own
    min_df.
    
    It also returns x and y, so that x can be used to fit the vectorizer, so that the different models can still
    predict based on words that aren't necessarily in its set of the data.
    
    X SHOULD NEVER TOUCH THE CLASSIFIER, ONLY THE VECTORIZER
    """
    # This section splits between train and test sets
    y='success'
    i, e = disjoint_sets(len(df), proportion)
    train = df[i]
    test = df[e]
    
    # Get X and Y
    x = df[field]
    y = df['success']
    
    # Get the training and test data for the x and y
    xtrain = train[field]
    xtest_final = test[field]
    ytrain = train['success']
    ytest_final = test['success']
    
    return x, y, xtrain, ytrain, xtest_final, ytest_final

def make_xy(x, y, vectorizer=None):
    """
    This function vecotrizes some x, and returns x and y
    """
    text = x
    
    if vectorizer == None:
        vectorizer = CountVectorizer(min_df=0)

    vectorizer.fit(text)
    x = vectorizer.transform(text)
    #X = X.toarray()
    
    return x, y

# Utilities for plotting the predictions/calibration of a model
def ysigma(frac, n):
    # This just returns the sigma for y for plotting the calibration plpot
    return math.sqrt(frac*(1.0 - frac)/n)

def calibration_plot(clf, X, Y):
    """
    This shows the calibration plot for a classifier, taking the predictions, sorting them into groups based on
    predicted success rates, and then aggregating the actual success rate of that group
    """
    # Get the probabilities, set the number of bins
    nbins = 30
    proba = [row[1] for row in clf.predict_proba(X)]
    
    # make a dataframe for predicted probabilities and observed success, and sort them into bins
    df        = pd.DataFrame({"proba": proba, "label": Y})
    df["bin"] = df.apply(lambda x: math.floor(x['proba'] * nbins), axis=1)
    
    # split the dataframe into groups based on what section of the probability distribution it falls in
    # then, aggregate it to get the observed success rates
    bins      = df.groupby("bin")
    observed  = bins['label'].aggregate(np.mean).values
    predicted = bins['proba'].aggregate(np.mean).values
    bin_size  = bins['label'].aggregate(len).values
    
    # calculate sigmas, spread it out within two standard deviations
    sigmas    = [2*ysigma(observed[i], bin_size[i]) for i in range(len(bin_size))] # 95% confidence interval
    
    # plot the observed vs the predicted with error bars, and a line for perfect calibration
    plt.errorbar(predicted, observed, yerr=np.vstack([sigmas, sigmas]))
    plt.plot([0,1], [0,1], color="r")
    plt.show()
    
    # plot the distribution of probability assignments
    plt.hist(proba, color='orange')
    plt.show()

def train_score_display(vectorizer, clf, x, xtrain, ytrain, xtest, ytest, fit_params={}):
    """
    This function fits a model's vectorizer to all the data (so that it can handle arbitrary data input),
    fits the classifier to x values vectorized using that vectorizer, then does a calibration plot,
    returning the resulting vecotrizer and classifier
    
    Also, it never gets passed y so that no modification from within it lets it train on the whole set
    """
    # Fit and transform according to the vectorizer
    vectorizer.fit(x)
    xtrain_v = vectorizer.transform(xtrain)
    xtest_v = vectorizer.transform(xtest)
    
    # train it
    if 'weight' in fit_params:
        clf.fit(xtrain_v, ytrain, weight=fit_params['weight'])
    else:
        clf.fit(xtrain_v, ytrain)
    
    # score and plot it
    score = clf.score(xtest_v, ytest)
    print "Score: ", score
    calibration_plot(clf, xtest_v, ytest)
    
    return vectorizer, clf

# This is the code for the random exploration of the parameter space
def mult_dict(d, s):
    """This function multiplies every value in a dictionary by a scalar"""
    ans = {}
    for key in d.keys():
        ans[key] = d[key] * s
    return ans

def test_precision(params, sigmas, precision):
    """
    This function checks whether the 95% confidence interval around a set of parameters is within precision proportion
    of the current parameters.
    
    I.e. if precision = .1, it checks the 95% confidence interval falls within +- 5% of each parameter
    """
    for key in params.keys():
        target = params[key] * precision / 2.0
        if sigmas[key] * 2 > target:
            return False
    return True
        
def random_move(params, sigmas):
    """
    This function has a 50% chance of taking a normally distributed step from the current position along each parameter
    """
    ans = {}
    for key in params.keys():
        if random.random() < .5: # do we change this parameter?
            ans[key] = -1
            while ans[key] <= 0: # keep the parameter above zero
                ans[key] = params[key] + random.gauss(0, sigmas[key])
        else:
            ans[key] = params[key]
            
    if params == ans: # if no changes, try again
        return random_move(params, sigmas)
    return ans

def random_optimization(model, params, sigmas, xtrain, ytrain, precision = .1):
    """
    This function implements random_optimization, taking normally distributed steps along each parameter, and keeping
    the change if it's an improvement. When the current best position remains the best, decrease sigma to take
    smaller steps.
    """
    best_params = params
    i = 0
    best_score  = model(best_params, xtrain, ytrain) # look at the best score
    
    while not test_precision(best_params, sigmas, precision): # repeat until we achive the desired precision
        new_params = random_move(best_params, sigmas)    # take a step
        new_score  = model(new_params, xtrain, ytrain)   # score the new model
        
        if new_score > best_score:   # if it's better, trade for it
            best_params = new_params
            best_score  =  new_score
        else:                        # otherwise, decrease sigma
            sigmas = mult_dict(sigmas, .9)
        
        i += 1
        if i % 10 == 1:
            print i, best_params, best_score, sigmas
            
    return best_params, best_score

# These functions all make a vectorizer and classifier with the given parameters, then do a cv score on the training set
def nb_classifier(params, xtrain, ytrain):
    vectorizer = CountVectorizer(min_df = params['min_df'])
    vectorizer.fit(xtrain)
    X, Y = make_xy(xtrain, ytrain, vectorizer)
    clf = MultinomialNB(alpha = params['alpha'])
    return np.mean(cv.cross_val_score(clf, X, Y, scoring=log_likelihood))

def log_classifier(params, xtrain, ytrain):
    vectorizer = CountVectorizer(min_df = params['min_df'])
    vectorizer.fit(xtrain)
    X, Y = make_xy(xtrain, ytrain, vectorizer)
    clf = sklm.LogisticRegression(C=params['C'])
    return np.mean(cv.cross_val_score(clf, X, Y, scoring=log_likelihood))

def dt_classifier(params, xtrain, ytrain):
    vectorizer = CountVectorizer(min_df = params['min_df'])
    vectorizer.fit(xtrain)
    X, Y = make_xy(xtrain, ytrain, vectorizer)
    clf =  sktr.DecisionTreeClassifier()
    return np.mean(cv.cross_val_score(clf, X.toarray(), Y, scoring=tree_cost))
```

```python
class CombinationFunction:    
    def __init__(self):
        pass
    
    def predict(self, parameters, prob_mat):
        return [self.combine(parameters, x) for x in prob_mat]
    
    def combine(self, parameters, x):
        raise NotImplementedError("Subclass must implement method")
        
    def gradient(self, parameters, x):
        raise NotImplementedError("Subclass must implement method")
        
    def dpdw(self, parameters, prob_mat):
        return np.matrix([self.gradient(parameters, row) for row in prob_mat])

class LinComb(CombinationFunction):
    def combine(self, parameters, probability_row):
        return sum([wi*xi for wi, xi in zip(parameters, probability_row)])
    
    def gradient(self, parameters, probability_row):
        return [probi for wi, probi in zip(parameters, probability_row)]

class LogComb(CombinationFunction):
    def combine(parameters, probability_row):
        return math.exp(linear_combination(parameters, probability_row))
    
    def gradient(self, parameters, probability_row):
        current = LinComb.combine(LinComb(), parameters, probability_row)
        return [probi * math.exp(current) for wi, probi in zip(parameters, probability_row)]
    
class MyComb(CombinationFunction):
    def combine(self, parameters, probability_row):
        return 1.0/(1.0 + -1*math.exp(LinComb.combine(LinComb(), parameters, [math.log(p/(1-p)) for p in probability_row])))
    
    def gradient(self, parameters, probability_row):
        current = self.combine(parameters, probability_row)
        return [-1*probi * math.exp(current) for wi, probi in zip(parameters, probability_row)]

def log_derivative(p, y):
    if y == 1:
        return 1.0 / p
    else:
        return -1.0 / (1 - p)

def ll_grad(pvector, ytrain):
    #print [log_derivative(p, y) for p,y in zip(pvector, ytrain)]
    return np.matrix([log_derivative(p, y) for p,y in zip(pvector, ytrain)])
    
def gradient(comb_func, parameters, prob_mat, ytrain):
    P    = comb_func.predict(parameters, prob_mat)
    dcdP = ll_grad(P, ytrain)
    dPdw = comb_func.dpdw(parameters, prob_mat)
    # print dcdP
    # print dPdw
    return dcdP * dPdw

def gradient_descent(mg, comb_func, parameters, xtrain, ytrain, gamma = .1):
    pm     = mg.prob_matrix(xtrain)
    params = np.matrix(parameters)
    grad   = gradient(comb_func, parameters, pm, ytrain)
    i      = 0
    
    while np.linalg.norm(grad) > 10:
        step   = (gamma / len(xtrain)) * grad  
        params += step 
        
        i += 1
        
        if i % 500 == 1:
            print i, ":  ", params, grad
            
        grad   = gradient(comb_func, params.tolist()[0], pm, ytrain)
    return params
```

```python
# get the predicted values from the original predictions and the weightings we found
def weighted_predict(pmat, weights, y):
    
    weighted_prediction = np.zeros(y.size)
        
    for col, weight in enumerate(weights):
        weighted_prediction += weight * pmat[:, col]
        
    #if anything is predicted at above one or below 0, send back into the unit interval
    for i, term in enumerate(weighted_prediction):
        if term < 0:
            weighted_prediction[i] = 0
        if term > 1:
            weighted_prediction[i] = 1
            
    print 'log_like: ' +  str(log_likelihood(weighted_prediction, y))
    print 'fraction-guessed: ' +  str(fraction_correct(weighted_prediction, y))
        
    return weighted_prediction

#cost functions
def log_likelihood(proba, y):
    L = 0
    
    #fudged slightly so that things don't get wrecked by zero-probability events (see explanation in markdown above)
    for i, term in enumerate(proba):
        proba[i] += 1./1000 if term == 0 else 0
        proba[i] -= 1./1000 if term == 1 else 0
        
    true_logs = np.log(proba)
    false_logs = np.log(1 - proba)
    
    for index, value in enumerate(y):
        L += true_logs[index] if value == 1 else false_logs[index]
            
    return L

def fraction_correct(proba, y):
    guess = np.around(proba)
    return np.mean(guess == y)
```
